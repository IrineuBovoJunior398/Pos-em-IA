```python
def calcular_media(numeros):
    return sum(numeros) / len(numeros)

media = calcular_media([10, 18, 20, 22, 29, 37, 43, 48, 54, 65, 77, 99, 121])
print(media)
```

```python
def mediana(numeros):
    """Calcula a mediana de uma lista de números."""
    nums_ordenados = sorted(numeros)
    n = len(nums_ordenados)
    if n % 2 == 0:
        # Par: média dos dois centrais
        return (nums_ordenados[n//2 - 1] + nums_ordenados[n//2]) / 2
    else:
        # Ímpar: valor central
        return nums_ordenados[n//2]

def desvio_padrao(numeros):
    """Calcula o desvio padrão populacional (dividido por N)."""
    if len(numeros) == 0:
        return 0
    media = sum(numeros) / len(numeros)
    variancia = sum((x - media)**2 for x in numeros) / len(numeros)
    return variancia ** 0.5  # Raiz quadrada da variância

# Exemplo de uso
dados = [10, 18, 20, 23, 29, 30, 32, 37, 40, 44, 52, 56, 60]
print(f"Mediana: {mediana(dados)}")  # Saída: 35.0
print(f"Desvio padrão: {desvio_padrao(dados):.2f}")  # Saída: 15.81
```
