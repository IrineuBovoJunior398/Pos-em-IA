```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
import numpy as np

# TransaГ§Гµes normais com 10 atributos (valor, hora, tipo, etc)
X = np.random.normal(0, 1, (1000, 10))

entrada = Input(shape=(10,))
codificado = Dense(4, activation='relu')(entrada)   # comprime pra 4
decodificado = Dense(10, activation='linear')(codificado)  # reconstrГіi os 10

autoencoder = Model(entrada, decodificado)
autoencoder.compile(optimizer='adam', loss='mse')
autoencoder.fit(X, X, epochs=20)  # entrada = saГ­da esperada

# TransaГ§ГЈo suspeita
suspeita = np.array([[10, -5, 3, 8, 1, 0, -2, 7, 4, 9]])
reconstruida = autoencoder.predict(suspeita)
erro = np.mean((suspeita - reconstruida) ** 2)
print(f"Erro de reconstruГ§ГЈo: {erro:.4f}")  # alto = suspeito
```
