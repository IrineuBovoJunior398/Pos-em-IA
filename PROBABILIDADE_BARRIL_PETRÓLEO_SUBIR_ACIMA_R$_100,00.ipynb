```python
# ============================================================
#  ANÁLISE DE PROBABILIDADE DO PETRÓLEO > $100
#  com Rede Neural + Contexto Geopolítico EUA-Irã
#  Autor: Inner AI Fusion | Data: Março 2026
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import warnings
warnings.filterwarnings('ignore')

# --- Dependências (instale se necessário) ---
# pip install yfinance scikit-learn tensorflow matplotlib seaborn

import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, LSTM, Reshape
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam

# Seed para reprodutibilidade
np.random.seed(42)
tf.random.set_seed(42)

print("=" * 60)
print("  REDE NEURAL: PROBABILIDADE DO PETRÓLEO SUPERAR $100")
print("  Contexto: Tensão Geopolítica EUA x IRÃ (2026)")
print("=" * 60)

# ============================================================
# BLOCO 1: COLETA DE DADOS REAIS (yfinance)
# ============================================================
print("\n[1/6] Coletando dados reais do mercado...")

# Ativos coletados
tickers = {
    'CL=F':  'Petróleo WTI (Futuro)',
    'BZ=F':  'Petróleo Brent (Futuro)',
    'GC=F':  'Ouro (Futuro)',
    'DX-Y.NYB': 'Dólar Index (DXY)',
    'USO':   'ETF Petróleo (USO)',
    'USDIOL=X': 'não usado',  # placeholder
}

# Download dos dados (últimos 3 anos para ter volume de treino)
data_raw = yf.download(
    ['CL=F', 'BZ=F', 'GC=F', 'DX-Y.NYB', 'USO'],
    start='2022-01-01',
    end=pd.Timestamp.today().strftime('%Y-%m-%d'),
    interval='1d',
    auto_adjust=True,
    progress=False
)

# Extraindo preços de fechamento
close_data = data_raw['Close'].copy()
close_data.columns = ['Brent', 'WTI', 'DXY', 'Ouro', 'USO']
close_data.dropna(inplace=True)

print(f"   Dados coletados: {len(close_data)} dias de pregão")
print(f"   Período: {close_data.index[0].date()} → {close_data.index[-1].date()}")
print(f"   Último preço WTI: ${close_data['WTI'].iloc[-1]:.2f}")
print(f"   Último preço Brent: ${close_data['Brent'].iloc[-1]:.2f}")

# ============================================================
# BLOCO 2: ENGENHARIA DE FEATURES (Indicadores Técnicos)
# ============================================================
print("\n[2/6] Engenharia de Features...")

df = close_data.copy()

# --- Médias Móveis ---
for w in [5, 10, 20, 50]:
    df[f'WTI_MA{w}'] = df['WTI'].rolling(w).mean()
    df[f'Brent_MA{w}'] = df['Brent'].rolling(w).mean()

# --- RSI (Relative Strength Index) ---
def calcular_rsi(series, periodo=14):
    delta = series.diff()
    gain = delta.where(delta > 0, 0).rolling(periodo).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(periodo).mean()
    rs = gain / (loss + 1e-9)
    return 100 - (100 / (1 + rs))

df['RSI_WTI'] = calcular_rsi(df['WTI'])
df['RSI_Brent'] = calcular_rsi(df['Brent'])

# --- Bandas de Bollinger ---
df['BB_medio'] = df['WTI'].rolling(20).mean()
df['BB_std'] = df['WTI'].rolling(20).std()
df['BB_superior'] = df['BB_medio'] + 2 * df['BB_std']
df['BB_inferior'] = df['BB_medio'] - 2 * df['BB_std']
df['BB_posicao'] = (df['WTI'] - df['BB_inferior']) / (df['BB_superior'] - df['BB_inferior'] + 1e-9)

# --- MACD ---
ema12 = df['WTI'].ewm(span=12).mean()
ema26 = df['WTI'].ewm(span=26).mean()
df['MACD'] = ema12 - ema26
df['MACD_signal'] = df['MACD'].ewm(span=9).mean()
df['MACD_hist'] = df['MACD'] - df['MACD_signal']

# --- Volatilidade e Retornos ---
df['Retorno_1d'] = df['WTI'].pct_change(1)
df['Retorno_5d'] = df['WTI'].pct_change(5)
df['Retorno_10d'] = df['WTI'].pct_change(10)
df['Volatilidade_20d'] = df['Retorno_1d'].rolling(20).std() * np.sqrt(252)

# --- Spread Brent-WTI ---
df['Spread_BW'] = df['Brent'] - df['WTI']

# --- Correlação Dólar (inversa ao petróleo) ---
df['DXY_retorno'] = df['DXY'].pct_change(1)
df['Ouro_retorno'] = df['Ouro'].pct_change(1)

# --- Distância do nível $100 ---
df['Dist_100'] = (100 - df['WTI']) / df['WTI']
df['Perc_de_100'] = df['WTI'] / 100

# --- Momentum ---
df['Momentum_5'] = df['WTI'] / df['WTI'].shift(5) - 1
df['Momentum_20'] = df['WTI'] / df['WTI'].shift(20) - 1

# ============================================================
# BLOCO 3: SIMULAÇÃO DE ÍNDICE GEOPOLÍTICO EUA-IRÃ
# ============================================================
print("\n[3/6] Incorporando Fator Geopolítico EUA x Irã...")

# Eventos geopolíticos conhecidos que impactaram o petróleo
# (datas reais com pesos de impacto estimados)
eventos_geo = {
    '2022-02-24': 1.0,   # Invasão da Ucrânia
    '2022-10-05': 0.7,   # Corte OPEC+
    '2023-10-07': 0.9,   # Conflito Hamas-Israel (tensão regional)
    '2023-12-01': 0.6,   # Ataques Houthis no Mar Vermelho
    '2024-04-01': 0.8,   # Ataque Irã a Israel (direto)
    '2024-04-14': 0.9,   # Resposta Israel ao Irã
    '2024-07-30': 0.7,   # Escalada no Oriente Médio
    '2025-01-20': 0.5,   # Mudança governo EUA
    '2025-06-01': 0.8,   # Sanções EUA ao Irã (estimado)
    '2025-09-15': 0.9,   # Tensão no Estreito de Ormuz (estimado)
    '2026-01-01': 0.7,   # Início confronto EUA-Irã (estimado)
    '2026-02-15': 0.85,  # Escalada do conflito (estimado)
    '2026-03-01': 0.95,  # Prorrogação da guerra (contexto atual)
}

# Criar série temporal de índice geopolítico com decaimento
df['Geo_Index'] = 0.3  # Linha de base (sempre há alguma tensão)

for data_evento, intensidade in eventos_geo.items():
    try:
        data_dt = pd.Timestamp(data_evento)
        if data_dt in df.index or data_dt < df.index[-1]:
            idx_pos = df.index.searchsorted(data_dt)
            if idx_pos < len(df):
                for i in range(idx_pos, min(idx_pos + 30, len(df))):
                    decaimento = intensidade * np.exp(-0.1 * (i - idx_pos))
                    df['Geo_Index'].iloc[i] = max(
                        df['Geo_Index'].iloc[i], decaimento
                    )
    except:
        pass

# Suavizar o índice geopolítico
df['Geo_Index_smooth'] = df['Geo_Index'].rolling(5).mean().fillna(df['Geo_Index'])

# Fator de risco de disrupção de oferta (Estreito de Ormuz)
df['Risco_Ormuz'] = df['Geo_Index_smooth'] * 0.4  # 40% do petróleo passa por lá

print(f"   Índice Geopolítico atual: {df['Geo_Index_smooth'].iloc[-1]:.3f}")
print(f"   Risco de disrupção (Ormuz): {df['Risco_Ormuz'].iloc[-1]:.3f}")

# ============================================================
# BLOCO 4: VARIÁVEL ALVO E PREPARAÇÃO DOS DADOS
# ============================================================
print("\n[4/6] Preparando Dataset para a Rede Neural...")

# Variável alvo: O preço do WTI vai superar $100 nos próximos N dias?
HORIZONTE = 7  # dias à frente

df['Target'] = (df['WTI'].shift(-HORIZONTE) > 100).astype(int)

# Features selecionadas
features = [
    'WTI', 'Brent', 'Ouro', 'DXY',
    'WTI_MA5', 'WTI_MA10', 'WTI_MA20', 'WTI_MA50',
    'RSI_WTI', 'RSI_Brent',
    'BB_posicao', 'MACD', 'MACD_hist',
    'Retorno_1d', 'Retorno_5d', 'Retorno_10d',
    'Volatilidade_20d', 'Spread_BW',
    'DXY_retorno', 'Ouro_retorno',
    'Dist_100', 'Perc_de_100',
    'Momentum_5', 'Momentum_20',
    'Geo_Index_smooth', 'Risco_Ormuz'
]

# Remover NaN
df_model = df[features + ['Target']].dropna()

print(f"   Total de amostras: {len(df_model)}")
print(f"   Features utilizadas: {len(features)}")
print(f"   Horizonte de previsão: {HORIZONTE} dias")
print(f"   Proporção classe 1 (>$100): {df_model['Target'].mean():.2%}")

X = df_model[features].values
y = df_model['Target'].values

# Normalização
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Split temporal (não aleatório — importante para séries temporais!)
split = int(len(X_scaled) * 0.80)
X_train = X_scaled[:split]
X_test  = X_scaled[split:]
y_train = y[:split]
y_test  = y[split:]

print(f"   Treino: {len(X_train)} | Teste: {len(X_test)}")

# ============================================================
# BLOCO 5: CONSTRUÇÃO E TREINAMENTO DA REDE NEURAL
# ============================================================
print("\n[5/6] Construindo e Treinando a Rede Neural...")

def construir_modelo(n_features, taxa_aprendizado=0.001):
    """
    Rede Neural Densa com arquitetura não-linear profunda.
    Usa BatchNorm + Dropout para regularização.
    """
    modelo = Sequential([
        # Camada de entrada
        Dense(128, activation='relu', input_shape=(n_features,),
              kernel_regularizer=tf.keras.regularizers.l2(0.001)),
        BatchNormalization(),
        Dropout(0.3),

        # Camada oculta 1 — não-linear profunda
        Dense(256, activation='relu',
              kernel_regularizer=tf.keras.regularizers.l2(0.001)),
        BatchNormalization(),
        Dropout(0.4),

        # Camada oculta 2
        Dense(128, activation='relu',
              kernel_regularizer=tf.keras.regularizers.l2(0.001)),
        BatchNormalization(),
        Dropout(0.3),

        # Camada oculta 3
        Dense(64, activation='relu',
              kernel_regularizer=tf.keras.regularizers.l2(0.001)),
        BatchNormalization(),
        Dropout(0.2),

        # Camada oculta 4 — captura relações residuais
        Dense(32, activation='relu'),
        Dropout(0.1),

        # Saída — probabilidade (sigmoid)
        Dense(1, activation='sigmoid')
    ])

    modelo.compile(
        optimizer=Adam(learning_rate=taxa_aprendizado),
        loss='binary_crossentropy',
        metrics=['accuracy',
                 tf.keras.metrics.AUC(name='auc'),
                 tf.keras.metrics.Precision(name='precision'),
                 tf.keras.metrics.Recall(name='recall')]
    )
    return modelo

modelo = construir_modelo(len(features))
modelo.summary()

# Callbacks
early_stop = EarlyStopping(
    monitor='val_auc', mode='max',
    patience=20, restore_best_weights=True, verbose=1
)
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss', factor=0.5,
    patience=10, min_lr=1e-6, verbose=1
)

# Peso para lidar com desbalanceamento de classes
peso_positivo = (y_train == 0).sum() / (y_train == 1).sum() if (y_train == 1).sum() > 0 else 1
class_weight = {0: 1.0, 1: float(peso_positivo)}
print(f"   Peso da classe positiva: {peso_positivo:.2f}")

# Treinamento
historico = modelo.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=200,
    batch_size=32,
    callbacks=[early_stop, reduce_lr],
    class_weight=class_weight,
    verbose=1
)

# ============================================================
# BLOCO 6: AVALIAÇÃO E RESULTADO FINAL
# ============================================================
print("\n[6/6] Avaliando o Modelo e Gerando Resultado...")

# Métricas no conjunto de teste
loss, acc, auc, prec, rec = modelo.evaluate(X_test, y_test, verbose=0)
print(f"\n  Acurácia no Teste: {acc:.2%}")
print(f"  AUC-ROC no Teste:  {auc:.4f}")
print(f"  Precisão:          {prec:.2%}")
print(f"  Recall:            {rec:.2%}")

# Predições
y_pred_prob = modelo.predict(X_test, verbose=0).flatten()
y_pred = (y_pred_prob > 0.5).astype(int)

print("\n  Relatório de Classificação:")
print(classification_report(y_test, y_pred, target_names=['< $100', '> $100']))

# --- Probabilidade ATUAL (dados mais recentes) ---
X_atual = scaler.transform(df_model[features].tail(1).values)
prob_atual = float(modelo.predict(X_atual, verbose=0).flatten()[0])

# Ajuste pelo índice geopolítico atual
geo_boost = df['Geo_Index_smooth'].iloc[-1] * 0.15
prob_ajustada = min(prob_atual + geo_boost, 0.99)

preco_atual_wti = df['WTI'].iloc[-1]
preco_atual_brent = df['Brent'].iloc[-1]

# Cotação estimada do dólar (buscar via yfinance)
try:
    dolar = yf.download('BRL=X', period='1d', progress=False)
    cotacao_dolar = float(dolar['Close'].iloc[-1])
except:
    cotacao_dolar = 5.85  # fallback estimado

preco_wti_brl = preco_atual_wti * cotacao_dolar
preco_brent_brl = preco_atual_brent * cotacao_dolar
meta_brl = 100 * cotacao_dolar

print("\n" + "=" * 60)
print("   RESULTADO DA ANÁLISE")
print("=" * 60)
print(f"   Data da análise:        {pd.Timestamp.today().strftime('%d/%m/%Y')}")
print(f"   Preço WTI atual:        ${preco_atual_wti:.2f}  (R$ {preco_wti_brl:.2f})")
print(f"   Preço Brent atual:      ${preco_atual_brent:.2f}  (R$ {preco_brent_brl:.2f})")
print(f"   Meta de análise:        $100.00 (R$ {meta_brl:.2f})")
print(f"   Cotação USD/BRL:        R$ {cotacao_dolar:.2f}")
print(f"   Índice Geopolítico:     {df['Geo_Index_smooth'].iloc[-1]:.3f} / 1.000")
print(f"   Horizonte de previsão:  {HORIZONTE} dias úteis")
print(f"   Prob. base (modelo):    {prob_atual:.2%}")
print(f"   Boost geopolítico:      +{geo_boost:.2%}")
print(f"   PROBABILIDADE FINAL:    {prob_ajustada:.2%}")
print("=" * 60)

# Classificação do risco
if prob_ajustada >= 0.70:
    nivel = "ALTO"
    cor_txt = "⚠️  RISCO ELEVADO"
elif prob_ajustada >= 0.45:
    nivel = "MÉDIO"
    cor_txt = "⚡ RISCO MODERADO"
else:
    nivel = "BAIXO"
    cor_txt = "✅ RISCO BAIXO"

print(f"\n   {cor_txt} de o WTI superar $100 nos próximos {HORIZONTE} dias")
print(f"   Nível de risco: {nivel}")

# ============================================================
# BLOCO 7: VISUALIZAÇÕES
# ============================================================
fig, axes = plt.subplots(3, 2, figsize=(18, 16))
fig.suptitle('Análise de Probabilidade: Petróleo > $100\nContexto: Guerra EUA × Irã (2026)',
             fontsize=16, fontweight='bold', y=0.98)

# --- 1. Preço histórico com linha $100 ---
ax1 = axes[0, 0]
ax1.plot(df.index, df['WTI'], color='#1f77b4', linewidth=1.5, label='WTI')
ax1.plot(df.index, df['Brent'], color='#ff7f0e', linewidth=1.5, alpha=0.8, label='Brent')
ax1.axhline(y=100, color='red', linestyle='--', linewidth=2, label='Meta: $100')
ax1.fill_between(df.index, df['WTI'], 100,
                 where=(df['WTI'] > 100), alpha=0.3, color='green', label='Acima $100')
ax1.fill_between(df.index, df['WTI'], 100,
                 where=(df['WTI'] <= 100), alpha=0.1, color='red')
ax1.set_title('Preço Histórico do Petróleo (WTI & Brent)', fontweight='bold')
ax1.set_ylabel('USD / Barril')
ax1.legend(loc='upper right')
ax1.grid(True, alpha=0.3)
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
ax1.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
plt.setp(ax1.xaxis.get_majorticklabels(), rotation=45)

# --- 2. Índice Geopolítico ---
ax2 = axes[0, 1]
ax2.fill_between(df.index, df['Geo_Index_smooth'], alpha=0.6, color='darkred', label='Índice Geopolítico')
ax2.plot(df.index, df['Geo_Index_smooth'], color='red', linewidth=1.5)
ax2.axhline(y=0.7, color='orange', linestyle='--', alpha=0.8, label='Nível de Alerta')
ax2.axhline(y=0.9, color='darkred', linestyle='--', alpha=0.8, label='Nível Crítico')
ax2.set_title('Índice de Risco Geopolítico EUA × Irã', fontweight='bold')
ax2.set_ylabel('Intensidade (0-1)')
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
ax2.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
plt.setp(ax2.xaxis.get_majorticklabels(), rotation=45)

# --- 3. Curva de Aprendizado ---
ax3 = axes[1, 0]
ep = range(1, len(historico.history['loss']) + 1)
ax3.plot(ep, historico.history['loss'], label='Loss Treino', color='blue')
ax3.plot(ep, historico.history['val_loss'], label='Loss Validação', color='orange')
ax3.set_title('Curva de Aprendizado (Loss)', fontweight='bold')
ax3.set_xlabel('Época')
ax3.set_ylabel('Binary Crossentropy')
ax3.legend()
ax3.grid(True, alpha=0.3)

# --- 4. AUC ao longo do treinamento ---
ax4 = axes[1, 1]
ax4.plot(ep, historico.history['auc'], label='AUC Treino', color='green')
ax4.plot(ep, historico.history['val_auc'], label='AUC Validação', color='red')
ax4.axhline(y=0.7, color='gray', linestyle='--', alpha=0.5, label='Threshold 0.7')
ax4.set_title('AUC-ROC ao Longo do Treinamento', fontweight='bold')
ax4.set_xlabel('Época')
ax4.set_ylabel('AUC')
ax4.legend()
ax4.grid(True, alpha=0.3)

# --- 5. Matriz de Confusão ---
ax5 = axes[2, 0]
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax5,
            xticklabels=['< $100', '> $100'],
            yticklabels=['< $100', '> $100'])
ax5.set_title('Matriz de Confusão (Conjunto de Teste)', fontweight='bold')
ax5.set_xlabel('Predição')
ax5.set_ylabel('Real')

# --- 6. Painel de Resultado Final ---
ax6 = axes[2, 1]
ax6.axis('off')

# Gauge visual (barra de probabilidade)
cor_gauge = '#e74c3c' if prob_ajustada >= 0.7 else '#f39c12' if prob_ajustada >= 0.45 else '#27ae60'

barra = ax6.barh(['Probabilidade'], [prob_ajustada * 100],
                  color=cor_gauge, height=0.4, alpha=0.85)
ax6.barh(['Probabilidade'], [100], color='#ecf0f1', height=0.4, alpha=0.5, zorder=0)
ax6.axvline(x=50, color='gray', linestyle='--', alpha=0.5)
ax6.axvline(x=70, color='orange', linestyle='--', alpha=0.5)
ax6.set_xlim(0, 100)

texto_resultado = (
    f"RESULTADO FINAL\n\n"
    f"WTI Atual: ${preco_atual_wti:.2f} (R$ {preco_wti_brl:.2f})\n"
    f"Brent Atual: ${preco_atual_brent:.2f} (R$ {preco_brent_brl:.2f})\n"
    f"Meta: $100.00 (R$ {meta_brl:.2f})\n\n"
    f"Prob. Base (RNA): {prob_atual:.1%}\n"
    f"Boost Geopolítico: +{geo_boost:.1%}\n"
    f"PROBABILIDADE FINAL: {prob_ajustada:.1%}\n\n"
    f"Risco: {nivel}\n"
    f"Horizonte: {HORIZONTE} dias úteis\n"
    f"Índice Geo: {df['Geo_Index_smooth'].iloc[-1]:.3f}/1.000"
)

ax6.text(50, -0.6, texto_resultado, ha='center', va='top',
         fontsize=11, fontweight='bold',
         bbox=dict(boxstyle='round,pad=0.8', facecolor='#f8f9fa',
                   edgecolor=cor_gauge, linewidth=2),
         transform=ax6.transData)

ax6.set_title(f'Probabilidade: WTI > $100 em {HORIZONTE} dias',
              fontweight='bold', fontsize=12, color=cor_gauge)

plt.tight_layout()
plt.savefig('petroleo_rna_analise.png', dpi=150, bbox_inches='tight',
            facecolor='white')
plt.show()
print("\n  Gráfico salvo: petroleo_rna_analise.png")

# ============================================================
# BLOCO 8: ANÁLISE DE SENSIBILIDADE (What-if)
# ============================================================
print("\n" + "=" * 60)
print("  ANÁLISE DE SENSIBILIDADE — CENÁRIOS")
print("=" * 60)

cenarios = {
    'Base (atual)':           {'geo': df['Geo_Index_smooth'].iloc[-1]},
    'Escalada leve (+10%)':   {'geo': min(df['Geo_Index_smooth'].iloc[-1] * 1.1, 1.0)},
    'Conflito aberto (+25%)': {'geo': min(df['Geo_Index_smooth'].iloc[-1] * 1.25, 1.0)},
    'Ormuz bloqueado':        {'geo': 0.98},
    'Cessar-fogo (-30%)':     {'geo': df['Geo_Index_smooth'].iloc[-1] * 0.7},
    'Acordo de paz (-50%)':   {'geo': df['Geo_Index_smooth'].iloc[-1] * 0.5},
}

print(f"\n  {'Cenário':<30} {'Índice Geo':>12} {'Probabilidade':>15}")
print(f"  {'-'*57}")

for nome, params in cenarios.items():
    boost_cen = params['geo'] * 0.15
    prob_cen = min(prob_atual + boost_cen, 0.99)
    marcador = " ◄ ATUAL" if nome == 'Base (atual)' else ""
    print(f"  {nome:<30} {params['geo']:>12.3f} {prob_cen:>14.2%}{marcador}")

print("\n" + "=" * 60)
print("  AVISO LEGAL")
print("=" * 60)
print("""
  Este código tem finalidade EDUCACIONAL e de pesquisa.
  NÃO constitui recomendação de investimento.

  Mercados financeiros são influenciados por fatores
  imprevisíveis. Modelos de ML têm limitações inerentes
  quando aplicados a séries temporais financeiras não-estacionárias.

  Consulte sempre um analista financeiro certificado
  antes de tomar decisões de investimento.
""")
print("=" * 60)
```
