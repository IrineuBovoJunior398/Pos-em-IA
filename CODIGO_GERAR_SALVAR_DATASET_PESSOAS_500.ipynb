```python

import numpy as np
import pandas as pd

# ==========================================
# Configurações
# ==========================================
# Se estiver no Colab, essa importação é para permitir o download
try:
    from google.colab import files
    COLAB = True
except ImportError:
    COLAB = False

# ==========================================
# 1. Gerar dataset de 500 pessoas
# ==========================================
np.random.seed(42)
n_pessoas = 500

# Tempo de tela (0 a 12 horas)
tempo_tela = np.random.uniform(0, 12, size=n_pessoas)

# Idade (12 a 70 anos)
idade = np.random.randint(12, 71, size=n_pessoas)

# Geração
geracao = []
for a in idade:
    if a <= 25:
        geracao.append(1)  # Gen Z
    elif a <= 40:
        geracao.append(2)  # Millennial
    elif a <= 55:
        geracao.append(3)  # Gen X
    else:
        geracao.append(4)  # Boomer
geracao = np.array(geracao)

# Frequência de culto (0 a 4 por semana)
frequencia_culto = np.random.randint(0, 5, size=n_pessoas)

# Frequência de pequeno grupo (0 a 2 por semana)
frequencia_pg = np.random.randint(0, 3, size=n_pessoas)

# Índices derivados
tempo_tela_norm = (tempo_tela / 12.0) * 10.0  # 0–10

peso_geracao = {1: 1.2, 2: 1.0, 3: 0.8, 4: 0.6}
peso_g = np.array([peso_geracao[g] for g in geracao])

indice_exposicao_digital = np.clip(tempo_tela_norm * peso_g, 0, 10)

comunhao_bruto = frequencia_culto + frequencia_pg  # 0–6
indice_comunhao = np.clip((comunhao_bruto / 6.0) * 10.0, 0, 10)

# Alvos (risco) – simulados de forma coerente
ruido_ans = np.random.normal(0, 1.0, size=n_pessoas)
ruido_dep = np.random.normal(0, 1.0, size=n_pessoas)

risco_ansiedade = np.clip(
    0.6 * indice_exposicao_digital - 0.4 * indice_comunhao + 5 + ruido_ans,
    0,
    10
)

risco_depressao = np.clip(
    0.5 * indice_exposicao_digital - 0.5 * indice_comunhao + 5 + ruido_dep,
    0,
    10
)

# Montar DataFrame
df_pessoas = pd.DataFrame({
    "ID_Pessoa": np.arange(1, n_pessoas + 1),
    "Tempo_Tela_Diario": np.round(tempo_tela, 3),
    "Idade": idade,
    "Geracao": geracao,
    "Frequencia_Culto_Semana": frequencia_culto,
    "Frequencia_Pequeno_Grupo": frequencia_pg,
    "Indice_Exposicao_Digital": np.round(indice_exposicao_digital, 3),
    "Indice_Comunhao": np.round(indice_comunhao, 3),
    "Risco_Ansiedade": np.round(risco_ansiedade, 3),
    "Risco_Depressao": np.round(risco_depressao, 3),
})

print("Amostra do dataset:")
print(df_pessoas.head())

# ==========================================
# 2. Salvar para CSV
# ==========================================
nome_arquivo = "dataset_pessoas_500.csv"

# Usar sep="," (padrão CSV) – ótimo para Colab e Python.
df_pessoas.to_csv("dataset_pessoas_500.csv", index=False)

print(f"\nArquivo '{"dataset_pessoas_500.csv"}' gerado com sucesso na pasta atual.")

# ==========================================
# 3. Fazer download para o computador (se estiver no Colab)
# ==========================================
if COLAB:
    print("Iniciando download do arquivo...")
    files.download("dataset_pessoas_500.csv")
else:
    print("Você não está no Colab; o arquivo foi salvo localmente na pasta do script.")
```
