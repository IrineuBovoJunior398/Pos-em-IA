```python
import numpy as np

# Dados originais (não-linearmente separáveis - círculos concêntricos)
# Estes são pontos simulados para dois círculos: interno (rótulo 0) e externo (rótulo 1)
X_original = np.array([[1, 1], [2, 2], [3, 3], [4, 4]])
y = [0, 0, 1, 1]  # Classes: 0 para círculo interno, 1 para externo

print("=== DADOS ORIGINAIS ===")
print("Coordenadas (x, y):")
print(X_original)
print("Rótulos (classes):", y)
print()

# Transformação para espaço de maior dimensão
# Adicionamos uma feature não-linear: o raio ao quadrado (x² + y²)
X_transformado = []
for x, y_coord in X_original:
    # Calcula o raio ao quadrado
    raio_quadrado = x**2 + y_coord**2
    # Adiciona ao novo array: [x, y, raio_quadrado]
    X_transformado.append([x, y_coord, raio_quadrado])

# Converte para array numpy para facilitar manipulação
X_transformado = np.array(X_transformado)

print("=== DADOS TRANSFORMADOS ===")
print("Coordenadas (x, y, raio²):")
print(X_transformado)
print("Rótulos permanecem os mesmos:", y)
print()

# Demonstração visual (opcional - requer matplotlib)
try:
    import matplotlib.pyplot as plt

    plt.figure(figsize=(10, 4))

    # Plot dos dados originais
    plt.subplot(1, 2, 1)
    plt.scatter(X_original[y==0, 0], X_original[y==0, 1], color='blue', label='Classe 0 (interno)')
    plt.scatter(X_original[y==1, 0], X_original[y==1, 1], color='red', label='Classe 1 (externo)')
    plt.title('Dados Originais\n(Não Linearmente Separáveis)')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.axis('equal')

    # Plot dos dados transformados (projetando em 2D para visualização)
    plt.subplot(1, 2, 2)
    plt.scatter(X_transformado[y==0, 2], X_transformado[y==0, 0], color='blue', label='Classe 0')
    plt.scatter(X_transformado[y==1, 2], X_transformado[y==1, 0], color='red', label='Classe 1')
    plt.title('Dados Transformados\n(Linearmente Separáveis)')
    plt.xlabel('Raio²')
    plt.ylabel('X')
    plt.legend()

    plt.tight_layout()
    plt.show()

except ImportError:
    print("Matplotlib não instalado - pulando visualização")

print("=== EXPLICAÇÃO ===")
print("Por que o código anterior não produzia saída?")
print("- Ele apenas definia variáveis e listas")
print("- Não havia comandos print() ou plt.show() para exibir resultados")
print("- Python executa o código, mas só mostra o que você manda imprimir")
print()
print("Por que a transformação funciona?")
print("- Dados originais: pontos (x,y) de círculos concêntricos não são separáveis por linha reta")
print("- Após transformação: adicionamos x²+y² (raio ao quadrado)")
print("- Agora no espaço 3D (x,y,raio²), as classes ficam separáveis por um plano")
print("- Classe 0 (interno): raio² pequeno, Classe 1 (externo): raio² grande")
```
