```python
!pip install schedule
import schedule
import time
from datetime import datetime

def tarefa_agendada():
    """Função que executa a automação da empresa (ex.: verificar agendamentos, enviar lembrete)."""
    agora = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    print(f"Tarefa executada automaticamente em {agora}")
    # Aqui você insere a lógica da empresa, ex.: conectar ao banco ou API
    # Exemplo simples: print de agendamento
    print("Agendamento automático: Enviando lembrete para clientes...")

# Agendamentos exemplos
schedule.every(1).hours.do(tarefa_agendada)  # A cada hora
schedule.every().day.at("09:00").do(tarefa_agendada)  # Todo dia às 09:00[1]
schedule.every(5).seconds.do(tarefa_agendada)  # Teste: a cada 5s

# Loop principal para executar agendamentos
while True:
    schedule.run_pending()
    time.sleep(1)
```
