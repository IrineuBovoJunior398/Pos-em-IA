```python
# ============================================================
# 1) INSTALAÇÃO DAS BIBLIOTECAS
# ============================================================
!pip install schedule joblib scikit-learn pandas numpy -q

# ============================================================
# 2) IMPORTAÇÃO DAS BIBLIOTECAS
# ============================================================
import os
import time
import schedule
import joblib
import pandas as pd
import numpy as np

from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier

# ============================================================
# 3) TREINAR E SALVAR UM MODELO DE EXEMPLO
#    (caso você ainda não tenha o arquivo .pkl)
# ============================================================
X, y = load_iris(return_X_y=True)

modelo_treino = RandomForestClassifier(random_state=42)
modelo_treino.fit(X, y)

joblib.dump(modelo_treino, "modelo_classificacao.pkl")

print("Modelo salvo com sucesso:", os.path.exists("modelo_classificacao.pkl"))

# ============================================================
# 4) CARREGAR O MODELO TREINADO
# ============================================================
modelo = joblib.load("modelo_classificacao.pkl")
print("Modelo carregado com sucesso.")

# ============================================================
# 5) FUNÇÃO PARA COLETAR DADOS NOVOS
#    Aqui estou usando um exemplo compatível com o Iris
# ============================================================
def coletar_dados():
    # Exemplo de nova amostra com 4 características
    # mesma estrutura usada no treino do modelo Iris
    dados = pd.DataFrame([[5.1, 3.5, 1.4, 0.2]],
                         columns=["sepal length (cm)", "sepal width (cm)",
                                  "petal length (cm)", "petal width (cm)"])
    return dados

# ============================================================
# 6) FUNÇÃO PARA TOMAR AÇÃO APÓS A PREVISÃO
# ============================================================
def enviar_notificacao(resultado):
    print("Previsão realizada com sucesso.")
    print("Resultado previsto:", resultado)

# ============================================================
# 7) FUNÇÃO AUTOMÁTICA PRINCIPAL
# ============================================================
def tarefa_automatica():
    try:
        print("\nExecutando tarefa automática...")

        # 1. Coletar dados novos
        dados = coletar_dados()
        print("Dados coletados:")
        print(dados)

        # 2. Fazer previsão com o modelo
        resultado = modelo.predict(dados)

        # 3. Tomar ação
        enviar_notificacao(resultado)

    except Exception as e:
        print("Erro durante a execução da tarefa:", e)

# ============================================================
# 8) TESTE IMEDIATO DA FUNÇÃO
#    Isso ajuda a verificar se tudo está funcionando antes
#    do agendamento
# ============================================================
tarefa_automatica()

# ============================================================
# 9) AGENDAMENTO DA TAREFA
#    Vai rodar todo dia às 09:00
# ============================================================
schedule.every().day.at("09:00").do(tarefa_automatica)

print("\nAgendamento configurado para rodar todos os dias às 09:00.")

# ============================================================
# 10) LOOP DE EXECUÇÃO CONTÍNUA
# ============================================================
while True:
    schedule.run_pending()
    time.sleep(60)
```
