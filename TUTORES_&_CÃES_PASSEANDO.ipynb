```python
import matplotlib.pyplot as plt
import random
import requests
from io import BytesIO
from PIL import Image

class Tutor:
    def __init__(self, nome, email):
        self.nome = nome
        self.email = email

    def conversar(self, outro_tutor):
        print(f"{self.nome} está conversando com {outro_tutor.nome} sobre cuidados com cães.")

class Cao:
    def __init__(self, nome, raca, energia=100):
        self.nome = nome
        self.raca = raca
        self.energia = energia
        self.historico_energia = [energia]  # Para plotar ao longo do tempo

    def brincar(self):
        # Reduz energia aleatoriamente entre 5 e 20
        reducao = random.randint(5, 20)
        self.energia = max(0, self.energia - reducao)
        self.historico_energia.append(self.energia)
        print(f"{self.nome} brincou e perdeu {reducao} de energia. Energia atual: {self.energia}")

# Simulação do parque
print("Bem-vindo ao Parque para Cães!")

# Criando 4 tutores com seus cães
tutores = [
    Tutor("Ana", "ana@email.com"),
    Tutor("Bruno", "bruno@email.com"),
    Tutor("Carla", "carla@email.com"),
    Tutor("Diego", "diego@email.com")
]

caes = [
    Cao("Rex", "Labrador", 100),
    Cao("Bella", "Poodle", 100),
    Cao("Max", "Bulldog", 100),
    Cao("Luna", "Golden Retriever", 100)
]

print("\nTutores e cães chegando:")
for i, (tutor, cao) in enumerate(zip(tutores, caes)):
    print(f"{i+1}. {tutor.nome} ({tutor.email}) com {cao.nome} ({cao.raca})")

# Simulando brincadeiras dos cães (5 rodadas)
print("\nCães brincando:")
for rodada in range(5):
    print(f"\nRodada {rodada+1}:")
    for cao in caes:
        cao.brincar()

# Tutores conversando
print("\nTutores conversando:")
for i in range(len(tutores)):
    for j in range(i+1, len(tutores)):
        tutores[i].conversar(tutores[j])

# Status finais
print("\nStatus finais:")
for tutor, cao in zip(tutores, caes):
    print(f"{tutor.nome}: {cao.nome} ({cao.raca}) - Energia: {cao.energia}")

# Plotando níveis de energia com matplotlib
plt.figure(figsize=(10, 6))
for cao in caes:
    plt.plot(cao.historico_energia, label=f"{cao.nome} ({cao.raca})")
plt.title("Níveis de Energia dos Cães ao Longo do Tempo")
plt.xlabel("Rodadas")
plt.ylabel("Energia")
plt.legend()
plt.grid(True)
plt.show()

# 🎉 AGORA AS FOTOS: Grade 2x2 das cenas do parque (tutores + cães brincando)
print("\nEncerrando com as cenas visuais do parque! Carregando 4 imagens direto de URL.")
print("Dica DS: requests + BytesIO + PIL = pipeline visão computacional sem arquivos locais!")

# URLs das 4 imagens geradas (pessoas/tutores + cães)
urls = [
    'https://generated-images.adapta.one/juniorbovo.servo777%40gmail.com/019d2065-c8a4-74e9-a503-79f014ad8eb9/2026-03-24T15-13-07-374Z_On_a_pleasant_late_afternoon_in_a_dog_park_Four_o.png',  # Cena original
    'https://generated-images.adapta.one/juniorbovo.servo777%40gmail.com/019d2065-c8a4-74e9-a503-79f014ad8eb9/2026-03-24T15-14-36-917Z_Create_a_vibrant_variation_of_the_image_with_golde.png',  # Variação vibrante
    'https://generated-images.adapta.one/juniorbovo.servo777%40gmail.com/019d2065-c8a4-74e9-a503-79f014ad8eb9/2026-03-24T15-14-48-255Z_Variation_with_a_closer_view_to_the_dogs_chasing_e.png',  # Close-up cães
    'https://generated-images.adapta.one/juniorbovo.servo777%40gmail.com/019d2065-c8a4-74e9-a503-79f014ad8eb9/2026-03-24T15-14-48-694Z_Dusk_variation_with_park_lamps_turning_on_dogs_st.png'   # Anoitecer
]

# Grade 2x2 para exibir imagens
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('🏞️ Cenas Finais: Tutores e Cães no Parque!', fontsize=16, fontweight='bold')

for i, ax in enumerate(axes.flat):
    response = requests.get(urls[i])  # 1. Baixa bytes da URL
    img_bytes = BytesIO(response.content)  # 2. Buffer em memória
    img = Image.open(img_bytes)  # 3. Abre com PIL
    ax.imshow(img)  # 4. Plota no subplot
    ax.set_title(f'Cena {i+1}: Tutores + Cães Brincando', fontsize=12)
    ax.axis('off')  # Sem eixos

plt.tight_layout()
plt.show()
import matplotlib.pyplot as plt
import random
import requests
from io import BytesIO
from PIL import Image

class Tutor:
    def __init__(self, nome, email):
        self.nome = nome
        self.email = email

    def conversar(self, outro_tutor):
        print(f"{self.nome} está conversando com {outro_tutor.nome} sobre cuidados com cães.")

class Cao:
    def __init__(self, nome, raca, energia=100):
        self.nome = nome
        self.raca = raca
        self.energia = energia
        self.historico_energia = [energia]  # Para plotar ao longo do tempo

    def brincar(self):
        # Reduz energia aleatoriamente entre 5 e 20
        reducao = random.randint(5, 20)
        self.energia = max(0, self.energia - reducao)
        self.historico_energia.append(self.energia)
        print(f"{self.nome} brincou e perdeu {reducao} de energia. Energia atual: {self.energia}")

# Simulação do parque
print("Bem-vindo ao Parque para Cães!")

# Criando 4 tutores com seus cães
tutores = [
    Tutor("Ana", "ana@email.com"),
    Tutor("Bruno", "bruno@email.com"),
    Tutor("Carla", "carla@email.com"),
    Tutor("Diego", "diego@email.com")
]

caes = [
    Cao("Rex", "Labrador", 100),
    Cao("Bella", "Poodle", 100),
    Cao("Max", "Bulldog", 100),
    Cao("Luna", "Golden Retriever", 100)
]

print("\nTutores e cães chegando:")
for i, (tutor, cao) in enumerate(zip(tutores, caes)):
    print(f"{i+1}. {tutor.nome} ({tutor.email}) com {cao.nome} ({cao.raca})")

# Simulando brincadeiras dos cães (5 rodadas)
print("\nCães brincando:")
for rodada in range(5):
    print(f"\nRodada {rodada+1}:")
    for cao in caes:
        cao.brincar()

# Tutores conversando
print("\nTutores conversando:")
for i in range(len(tutores)):
    for j in range(i+1, len(tutores)):
        tutores[i].conversar(tutores[j])

# Status finais
print("\nStatus finais:")
for tutor, cao in zip(tutores, caes):
    print(f"{tutor.nome}: {cao.nome} ({cao.raca}) - Energia: {cao.energia}")

# Plotando níveis de energia com matplotlib
plt.figure(figsize=(10, 6))
for cao in caes:
    plt.plot(cao.historico_energia, label=f"{cao.nome} ({cao.raca})")
plt.title("Níveis de Energia dos Cães ao Longo do Tempo")
plt.xlabel("Rodadas")
plt.ylabel("Energia")
plt.legend()
plt.grid(True)
plt.show()

# 🎉 AGORA AS FOTOS: Grade 2x2 das cenas do parque (tutores + cães brincando)
print("\nEncerrando com as cenas visuais do parque! Carregando 4 imagens direto de URL.")
print("Dica DS: requests + BytesIO + PIL = pipeline visão computacional sem arquivos locais!")

# URLs das 4 imagens geradas (pessoas/tutores + cães)
urls = [
    'https://generated-images.adapta.one/juniorbovo.servo777%40gmail.com/019d2065-c8a4-74e9-a503-79f014ad8eb9/2026-03-24T15-13-07-374Z_On_a_pleasant_late_afternoon_in_a_dog_park_Four_o.png',  # Cena original
    'https://generated-images.adapta.one/juniorbovo.servo777%40gmail.com/019d2065-c8a4-74e9-a503-79f014ad8eb9/2026-03-24T15-14-36-917Z_Create_a_vibrant_variation_of_the_image_with_golde.png',  # Variação vibrante
    'https://generated-images.adapta.one/juniorbovo.servo777%40gmail.com/019d2065-c8a4-74e9-a503-79f014ad8eb9/2026-03-24T15-14-48-255Z_Variation_with_a_closer_view_to_the_dogs_chasing_e.png',  # Close-up cães
    'https://generated-images.adapta.one/juniorbovo.servo777%40gmail.com/019d2065-c8a4-74e9-a503-79f014ad8eb9/2026-03-24T15-14-48-694Z_Dusk_variation_with_park_lamps_turning_on_dogs_st.png'   # Anoitecer
]

# Grade 2x2 para exibir imagens
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('🏞️ Cenas Finais: Tutores e Cães no Parque!', fontsize=16, fontweight='bold')

for i, ax in enumerate(axes.flat):
    response = requests.get(urls[i])  # 1. Baixa bytes da URL
    img_bytes = BytesIO(response.content)  # 2. Buffer em memória
    img = Image.open(img_bytes)  # 3. Abre com PIL
    ax.imshow(img)  # 4. Plota no subplot
    ax.set_title(f'Cena {i+1}: Tutores + Cães Brincando', fontsize=12)
    ax.axis('off')  # Sem eixos

plt.tight_layout()
plt.show()

```
