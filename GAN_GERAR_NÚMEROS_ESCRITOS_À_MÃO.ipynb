```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

# Geradora: transforma ruído aleatório em dado "falso"
geradora = Sequential([
    Dense(64, activation='relu', input_shape=(10,)),
    Dense(28*28, activation='sigmoid')  # imagem 28x28
])

# Discriminadora: diz se é real ou falso
discriminadora = Sequential([
    Dense(64, activation='relu', input_shape=(28*28,)),
    Dense(1, activation='sigmoid')  # 1=real, 0=falso
])

discriminadora.compile(optimizer='adam', loss='binary_crossentropy')

# Treinamento simplificado
ruido = np.random.normal(0, 1, (5, 10))
imagem_falsa = geradora.predict(ruido)
print("Imagem gerada com shape:", imagem_falsa.shape)
```
