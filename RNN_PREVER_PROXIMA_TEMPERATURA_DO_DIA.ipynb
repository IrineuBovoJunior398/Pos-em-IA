```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
import numpy as np

# SequГЄncia de 5 dias de temperatura para prever o 6Вє
X = np.array([[[25],[26],[27],[28],[29]]])  # shape: (amostras, passos, features)
y = np.array([[30]])

model = Sequential([
    SimpleRNN(16, input_shape=(5, 1)),  # processa cada dia em sequГЄncia
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.fit(X, y, epochs=50)
```
